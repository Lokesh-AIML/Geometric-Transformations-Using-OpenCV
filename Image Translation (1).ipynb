# **Image Translation**

Write a Python program using OpenCV to perform image translation on the **"Given image"**.

The program should:

1. Import `cv2`, `numpy`, and `matplotlib.pyplot`.
2. Read the image **"given** in color mode.
3. Create a **translation matrix** that moves the image  
   **50 pixels to the right** and **80 pixels down**.
4. Apply the translation
5. Display the following using a **1 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Translated Image**


```python
# ---------------- IMAGE TRANSLATION ----------------

# Import
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread(r"C:\Users\waran\Downloads\baseball.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Translation matrix
rows, cols = img.shape[:2]

M = np.float32([
    [1, 0, 50],
    [0, 1, 80]
])

translated = cv2.warpAffine(img_rgb, M, (cols, rows))
```

```python
# Display the image using Matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(translated)
plt.title("Translated Image")
plt.axis("off")

plt.show()
```

# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\waran\Downloads\baseball.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rows, cols = img.shape[:2]
```

```python
# Horizontal Shearing
M1 = np.float32([
    [1, 0.5, 0],
    [0, 1, 0]
])

horizontal_shear = cv2.warpAffine(img_rgb, M1, (cols + 200, rows))
```

```python
# Vertical Shearing
M2 = np.float32([
    [1, 0, 0],
    [0.5, 1, 0]
])

vertical_shear = cv2.warpAffine(img_rgb, M2, (cols, rows + 200))
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(horizontal_shear)
plt.title("Horizontal Shear")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(vertical_shear)
plt.title("Vertical Shear")
plt.axis("off")

plt.show()
```

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\waran\Downloads\baseball.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

rows, cols = img.shape[:2]
center = (cols // 2, rows // 2)
```

```python
# Rotation by 45 degrees
M1 = cv2.getRotationMatrix2D(center, 45, 1)
rotate45 = cv2.warpAffine(img_rgb, M1, (cols, rows))
```

```python
# Rotation by 90 degrees
M2 = cv2.getRotationMatrix2D(center, 90, 1)
rotate90 = cv2.warpAffine(img_rgb, M2, (cols, rows))
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(rotate45)
plt.title("45 Degree Rotation")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(rotate90)
plt.title("90 Degree Rotation")
plt.axis("off")

plt.show()
```
